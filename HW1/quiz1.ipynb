```python
#%%
import numpy as np
import statsmodels.formula.api as smf
import pandas as pd
```

```python
n=1000
x=np.random.normal(0,1,n)
y=10 + 5*x + np.random.normal(0,1,n)
```

```python
# %%
reg_data=pd.DataFrame({'x':x,'y':y})
reg_model = smf.ols(formula='y~x',data=reg_data)
fitted_model = reg_model.fit()
```

```python
# %%
print(fitted_model.summary())
```

```python
# %%
from statsmodels.graphics.regressionplots import abline_plot
ax=reg_data.plot(x='x', y='y',kind='scatter')
abline_plot(model_results=fitted_model,ax=ax,color='red')
```

```python
# %%
import statsmodels.api as sm
x = sm.add_constant(x)
model = sm.OLS(y, x)
results = model.fit()
```

```python
#%%
import seaborn as sns
sns.regplot(x="x", y="y", data=reg_data,line_kws={"color": "red"})
```
